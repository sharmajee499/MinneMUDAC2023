```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")
```

```python
# Read the Data
df = pd.read_csv("https://media.githubusercontent.com/media/sharmajee499/MinneMUDAC2023/main/Data/GameLogs.csv",
                 parse_dates=[0])
```

```python
# Only use the data that is necessary for calculation
df1 = df[['Date','DayofWeek','HomeTeamGameNumber','HomeTeam','HomeTeamScore', 'VisitingTeamGameNumber',
                  'VisitingTeam','VistingTeamScore']]

#Extract the year from the 'Date'
df1['Year'] = df1.Date.dt.year
df1.head()
```

```python
# Column for Wining Team
df1['WiningTeam'] = df1.apply(lambda x: x['HomeTeam'] if x['HomeTeamScore']> x['VistingTeamScore'] else x['VisitingTeam'], axis=1)

# Column for Loosing Team
df1['LosingTeam'] = df1.apply(lambda x: x['HomeTeam'] if x['HomeTeamScore']< x['VistingTeamScore'] else x['VisitingTeam'], axis=1)
```

```python
df1.head()
```

```python
# Add the column to show the total games on that year
df1['TotalGames'] = df1.groupby(['Year'])['Year'].transform(len)
df1.head()
```

```python
# Games Won by the whole team in that year
df1['WinTeamGameWonYear'] = df1.groupby(['Year','WiningTeam'])['WiningTeam'].transform(len)

# Game Lost by the whole team in that year
df1['LostTeamGameLostYear'] = df1.groupby(['Year','LosingTeam'])['LosingTeam'].transform(len)
```

```python
df1.head(20)
```

```python
#Number of Games won till that game in that season
#(+1 because the cumulative starts with 0 however the team have already won that perticular game)
df1['CumulativeGameWon'] = df1.groupby(['Year', 'WiningTeam'])['WiningTeam'].cumcount()+1
```

```python
dft = pd.DataFrame()
dft['Date'] = df1['Date']
dft['Year'] = df1['Date'].dt.year
dft['Team1'] = df1['HomeTeam']
dft['Team2'] = df1['VisitingTeam']
dft['Team1Score'] = df1['HomeTeamScore']
dft['Team2Score'] = df1['VistingTeamScore']
```

```python
dft = pd.melt(dft, id_vars=['Date','Year', 'Team1Score', 'Team2Score'], value_vars=['Team1', 'Team2'], var_name='Team', value_name='TeamName')
```
